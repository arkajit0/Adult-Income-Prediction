## Introduction

### Here we will build a classification model to check the income scales. The dataset we  are using has total 15 columns where 14 are feature columns and 1 label column.

The description of different columns are as follows:-

- age: the age of an individual
○ Integer greater than 0
- workclass: a general term to represent the employment status of an individual
○ Private, Self­emp­not­inc, Self­emp­inc, Federal­gov, Local­gov, State­gov,
Without­pay, Never­worked.
- fnlwgt: final weight. In other words, this is the number of people the census believes
the entry represents..
○ Integer greater than 0
- education: the highest level of education achieved by an individual.
○ Bachelors, Some­college, 11th, HS­grad, Prof­school, Assoc­acdm, Assoc­voc,
9th, 7th­8th, 12th, Masters, 1st­4th, 10th, Doctorate, 5th­6th, Preschool.
- education­num: the highest level of education achieved in numerical form.
○ Integer greater than 0
- marital­status: marital status of an individual. Married­civ­spouse corresponds to a
civilian spouse while Married­AF­spouse is a spouse in the Armed Forces.
○ Married­civ­spouse, Divorced, Never­married, Separated, Widowed,
Married­spouse­absent, Married­AF­spouse.
- occupation: the general type of occupation of an individual
○ Tech­support, Craft­repair, Other­service, Sales, Exec­managerial,
Prof­specialty, Handlers­cleaners, Machine­op­inspct, Adm­clerical,
Farming­fishing, Transport­moving, Priv­house­serv, Protective­serv,
Armed­Forces.
- relationship: represents what this individual is relative to others. For example an
individual could be a Husband. Each entry only has one relationship attribute and is
somewhat redundant with marital status. We might not make use of this attribute at all
○ Wife, Own­child, Husband, Not­in­family, Other­relative, Unmarried.
- race: Descriptions of an individual’s race
○ White, Asian­Pac­Islander, Amer­Indian­Eskimo, Other, Black.
- sex: the biological sex of the individual
○ Male, Female
- capital­gain: capital gains for an individual
○ Integer greater than or equal to 0
- capital­loss: capital loss for an individual
○ Integer greater than or equal to 0
- hours­per­week: the hours an individual has reported to work per week
○ continuous.
- native­country: country of origin for an individual
○ United­States, Cambodia, England, Puerto­Rico, Canada, Germany,
Outlying­US(Guam­USVI­etc), India, Japan, Greece, South, China, Cuba, Iran,
Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal,
Ireland, France, Dominican­Republic, Laos, Ecuador, Taiwan, Haiti, Columbia,
Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El­Salvador,
Trinadad&Tobago, Peru, Hong, Holand­Netherlands.
- the label: whether or not an individual makes more than $50,000 annually.
○ <=50k, >50k


```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pylab
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score, roc_curve
sns.set_style('dark')
%matplotlib inline
```

```python
#loading the train data
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
train_set.head()
```

```python
#loading the test data
test_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test' , skiprows = 1, header = None) 
test_set.head()
```

```python
#Renaming the columns in train and test data
col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',  'marital_status', 'occupation','relationship', 'race', 'sex', 'capital_gain',  'capital_loss', 'hours_per_week', 'native_country', 'wage_class']
train_set.columns = col_labels
test_set.columns = col_labels
```

```python
#replacing all unknown values with NaN
train_set.replace(' ?', np.nan, inplace=True)
test_set.replace(' ?', np.nan, inplace=True)
```

```python
#checking value count of each target variable
test_set['wage_class'].value_counts()
```

```python
#meging train and test dataset
data = pd.concat([train_set, test_set], axis=0)
```

```python
#checking dimension of new dataframe
data.shape
```

```python
#replace . in wages
data['wage_class'] = data['wage_class'].str.replace('.', '')
data['wage_class'].value_counts()
```

```python
#replacing the target variable with binary
data['wage_class'] = data['wage_class'].apply(lambda x: 0 if x.strip()=='<=50K' else 1)
```

```python
#checking for imbalanced data
plt.figure(figsize=(5,5))
data['wage_class'].value_counts().plot(kind='bar', title='count')
```

our data is imbalanced

```python
#separate class count
class_count_0, class_count_1 = data['wage_class'].value_counts()

#separate class
class_0 = data[data['wage_class'] == 0]
class_1 = data[data['wage_class'] == 1]

print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
#taking random rows using over sampling
class_1_over = class_1.sample(class_count_0, replace=True)

#mearging class_0 and new class_1
test_over = pd.concat([class_0, class_1_over], axis=0)
print(test_over['wage_class'].value_counts())
test_over['wage_class'].value_counts().plot(kind='bar', title='count')
```

our data is balance now

```python
#filling null values for workclass
qualifications = test_over['education'].str.strip().unique()
for qualification in qualifications:
    total_count = test_over[test_over['education'].str.strip()==qualification].workclass.count()
    categories_list = test_over[test_over['education'].str.strip()==qualification]['workclass'].dropna().unique()
    p_list = [test_over.loc[(test_over['education'].str.strip()==qualification)&(test_over['workclass']==workclass)].\
              shape[0]/total_count for workclass in categories_list]
#     break
    test_over.loc[(test_over['education'].str.strip()==qualification)&(test_over.workclass.isna()), 'workclass'] = \
            pd.Series(test_over[(test_over['education'].str.strip()==qualification)&(test_over.workclass.isna())].workclass.\
                      apply(lambda x:x if type(x)==str else np.random.choice(categories_list, 1, p_list)[0]))
```

```python
def plot_data(df, feature):
    plt.figure(figsize=(20,25))
    plt.subplot(2,2,1)
    df[feature].hist()
    plt.subplot(2,2,2)
    stats.probplot(df[feature], dist='norm', plot=pylab)
```

```python
plot_data(test_over, 'fnlwgt')
```

```python
#using log transformation
test_over['fnlwgt_log'] = np.log(test_over['fnlwgt'])
print('fnlwgt_log = ', test_over['fnlwgt_log'].skew())
plot_data(test_over, 'fnlwgt_log')
```

```python
#using reciprocal transformation
test_over['fnlwgt_reciprocal'] = 1/(test_over['fnlwgt'])
print('fnlwgt_reciprocal = ', test_over['fnlwgt_reciprocal'].skew())
plot_data(test_over, 'fnlwgt_reciprocal')
```

```python
#using sqrt transformation
test_over['fnlwgt_sqrt'] = np.sqrt(test_over['fnlwgt'])
print('fnlwgt_sqrt = ', test_over['fnlwgt_sqrt'].skew())
plot_data(test_over, 'fnlwgt_sqrt')
```

```python
#using exponential transformation
test_over['fnlwgt_exp'] = (test_over['fnlwgt'])**(1/1.2)
print('fnlwgt_exp = ', test_over['fnlwgt_exp'].skew())
plot_data(test_over, 'fnlwgt_exp')
```

```python
#using boxcox transformation
test_over['fnlwgt_boxcox'], parameter = stats.boxcox(test_over['fnlwgt'])
print('fnlwgt_boxcox = ', test_over['fnlwgt_boxcox'].skew())
print('parameter = ', parameter)
plot_data(test_over, 'fnlwgt_boxcox')
```

## Model Building

```python
#splitting to features and target
X = data_final.drop(['wage_class'], axis=1)
y = data_final['wage_class']
```

```python
#scaling the data
X_scaled = sc.fit_transform(X)
```

```python
#splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
#creating instance of xgboost and fitting the data
model = xgb.XGBClassifier(objective='binary:logistic')
model.fit(X_train, y_train)
```

```python
#saving the model
pickle.dump(model, open('final_model_v1.pkl', 'wb'))
```

```python
#checking performance of model
print('train score = ', model.score(X_train, y_train))
print('test score = ', model.score(X_test, y_test))
y_pred_model = model.predict(X_test)
conf_mtx = confusion_matrix(y_test, y_pred_model)
sns.heatmap(conf_mtx, fmt='', annot=True, cmap='rainbow')
TP = conf_mtx[0][0]
FP = conf_mtx[0][1]
FN = conf_mtx[1][0]
TN = conf_mtx[1][1]
accuracy = (TP+TN)/(TP+TN+FP+FN)
recall = TP/(TP+FN)
precision = TP/(TP+FP)
f1_score = 2*precision*recall/(precision+recall)
print("accuracy = ", accuracy)
print("recall = ", recall)
print("precision = ",precision)
print("F1_score = ", f1_score)
```

```python
#classification report
print(classification_report(y_test, y_pred_model, target_names=['<=50k', '>50k']))
```

```python
#calculating area under curve
auc = roc_auc_score(y_test, y_pred_model)
print('auc = ',auc)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_model)
plt.plot(fpr, tpr, color='orange', label='ROC')
plt.plot([0, 1], [0, 1], color='blue', linestyle='--',label='ROC curve (area = %0.2f)' % auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
```

## Hyperparameter Tuning

```python
#initializing parameters
param_grid={
   
    'learning_rate':[1,0.5,0.1,0.01,0.001],
    'max_depth': [3,5,10,20],
    'n_estimators':[10,50,100,200]
    
}
```

```python
#finding best parameters with gridsearchcv
grid= GridSearchCV(estimator=xgb.XGBClassifier(objective='binary:logistic'),param_grid=param_grid, verbose=3, n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
#best score and paramter
print(grid.best_score_)
print(grid.best_params_)
```

```python
#creating instance of xgboost with new parameters and fitting the data
model = xgb.XGBClassifier(**grid.best_params_)
model.fit(X_train, y_train)
```

```python
#saving the model
pickle.dump(model, open('final_model_v2.pkl', 'wb'))
```

```python
#checking performance of model
print('train score = ', model.score(X_train, y_train))
print('test score = ', model.score(X_test, y_test))
y_pred_model = model.predict(X_test)
conf_mtx = confusion_matrix(y_test, y_pred_model)
sns.heatmap(conf_mtx, fmt='', annot=True, cmap='rainbow')
TP = conf_mtx[0][0]
FP = conf_mtx[0][1]
FN = conf_mtx[1][0]
TN = conf_mtx[1][1]
accuracy = (TP+TN)/(TP+TN+FP+FN)
recall = TP/(TP+FN)
precision = TP/(TP+FP)
f1_score = 2*precision*recall/(precision+recall)
print("accuracy = ", accuracy)
print("recall = ", recall)
print("precision = ",precision)
print("F1_score = ", f1_score)
```
